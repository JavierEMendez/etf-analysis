# Financial Analysis

```python
# Importing the necessary modules:
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing the data after it has been cleaned: 
IAU_csv = Path("Resources/IAUData.csv", parse_dates=True, index_col='DateTime')
IVV_csv = Path("Resources/IVVData.csv", parse_dates=True, index_col='DateTime')
IWF_csv = Path("Resources/IWFData.csv", parse_dates=True, index_col='DateTime')
LUBAX_csv = Path("Resources/LUBAXData.csv", parse_dates=True, index_col='DateTime')
VDC_csv = Path("Resources/VDCData.csv", parse_dates=True, index_col='DateTime')
VNQ_csv = Path("Resources/VNQData.csv", parse_dates=True, index_col='DateTime')
```

```python
# Reading the CSV's with Pandas:
IAU_df = pd.read_csv(IAU_csv)
IVV_df = pd.read_csv(IVV_csv)
IWF_df = pd.read_csv(IWF_csv)
LUBAX_df = pd.read_csv(LUBAX_csv)
VDC_df = pd.read_csv(VDC_csv)
VNQ_df = pd.read_csv(VNQ_csv)
```

```python
# Converting Date column to datetime for all the dataframes before contatinating them,
# as well as renaming the 'Close' column to the ticker.
IAU_df.set_index(pd.to_datetime(IAU_df['Date'], infer_datetime_format=True), inplace=True)
IAU_df.drop(columns=['Date'], inplace=True)
IAU_df = IAU_df.rename(columns={'Close': 'IAU'})
IAU_df.head()
```

```python
IVV_df.set_index(pd.to_datetime(IVV_df['Date'], infer_datetime_format=True), inplace=True)
IVV_df.drop(columns=['Date'], inplace=True)
IVV_df = IVV_df.rename(columns={'Close': 'IVV'})
IVV_df.head()
```

```python
IWF_df.set_index(pd.to_datetime(IWF_df['Date'], infer_datetime_format=True), inplace=True)
IWF_df.drop(columns=['Date'], inplace=True)
IWF_df = IWF_df.rename(columns={'Close': 'IWF'})
IWF_df.head()
```

```python
LUBAX_df.set_index(pd.to_datetime(LUBAX_df['Date'], infer_datetime_format=True), inplace=True)
LUBAX_df.drop(columns=['Date'], inplace=True)
LUBAX_df = LUBAX_df.rename(columns={'Close': 'LUBAX'})

LUBAX_df.head()
```

```python
VDC_df.set_index(pd.to_datetime(VDC_df['Date'], infer_datetime_format=True), inplace=True)
VDC_df.drop(columns=['Date'], inplace=True)
VDC_df = VDC_df.rename(columns={'Close': 'VDC'})

VDC_df.head()
```

```python
VNQ_df.set_index(pd.to_datetime(VNQ_df['Date'], infer_datetime_format=True), inplace=True)
VNQ_df.drop(columns=['Date'], inplace=True)
VNQ_df = VNQ_df.rename(columns={'Close': 'VNQ'})

VNQ_df.head()
```

```python
# Combining all the data into one dataframe we can analyze:
analysis_df = pd.concat([VDC_df,IWF_df,VNQ_df,IVV_df,IAU_df,LUBAX_df], axis="columns", join="inner")
no_lubax_df = pd.concat([VDC_df,IWF_df,VNQ_df,IVV_df,IAU_df], axis="columns", join="inner")
analysis_df
```

```python
# Creating a dataframe with daily returns:
returns_df = analysis_df.pct_change()
returns_df
```

```python
# Plotting general returns:
returns_df.plot(figsize=(25,12.5))
```

```python
# Plotting cumulative returns:
cumulative_returns = (1 + returns_df).cumprod()
cumulative_returns.plot(figsize=(25,12.5))
```

```python
# Creating a box plot to show risk.
cumulative_returns.plot(kind="box", figsize=(20,10))
```

```python
# Calculating and plotting the rolling standard deviation for the ETFs:
analysis_df.rolling(21).std().plot(figsize=(25,12.5))
```

```python
# Constructing a correlation table:
correlation = analysis_df.corr()
correlation
```

```python
# Creating a heatmap of the correlation between all the ETFs.
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, cmap="RdBu")
```

```python
#Calculating a rolling Beta for each ETF:
# I am of the opinion that a simple beta does not tell a full
# story, and prefer rolling betas (even if they take longer to calculate).
VDC_rolling_covariance = VDC_df['VDC'].rolling(window=21).cov(VDC_df['VDC'])
VDC_rolling_variance = VDC_df['VDC'].rolling(window=21).var()
VDC_rolling_beta = VDC_rolling_covariance / VDC_rolling_variance
VDC_rolling_beta.plot(figsize=(20, 10), title='VDC Beta')
```

```python
# Rolling beta calculating for IWF
IWF_df_covariance = IWF_df['IWF'].rolling(window=21).cov(IWF_df['IWF'])
IWF_df_variance = IWF_df['IWF'].rolling(window=21).var()
IWF_df_beta = IWF_df_covariance / IWF_df_variance
IWF_df_beta.plot(figsize=(20, 10), title='IWF Beta')
```

```python
# Rolling beta calculating for VNQ
VNQ_covariance = VNQ_df['VNQ'].rolling(window=21).cov(VNQ_df['VNQ'])
VNQ_variance = VNQ_df['VNQ'].rolling(window=21).var()
VNQ_beta = VNQ_covariance / VNQ_variance
VNQ_beta.plot(figsize=(20, 10), title='VNQ Beta')
```

```python
# Rolling beta calculating for IVV
IVV_covariance = IVV_df['IVV'].rolling(window=21).cov(IVV_df['IVV'])
IVV_variance = IVV_df['IVV'].rolling(window=21).var()
IVV_beta = IVV_covariance / IVV_variance
IVV_beta.plot(figsize=(20, 10), title='IVV Beta')
```

```python
# Rolling beta calculating for IAU
IAU_covariance = IAU_df['IAU'].rolling(window=21).cov(IAU_df['IAU'])
IAU_variance = IAU_df['IAU'].rolling(window=21).var()
IAU_beta = IAU_covariance / IAU_variance
IAU_beta.plot(figsize=(20, 10), title='IAU Beta')
```

```python
# Rolling beta calculating for LUBAX
LUBAX_covariance = LUBAX_df['LUBAX'].rolling(window=21).cov(LUBAX_df['LUBAX'])
LUBAX_variance = LUBAX_df['LUBAX'].rolling(window=21).var()
LUBAX_beta = LUBAX_covariance / LUBAX_variance
LUBAX_beta.plot(figsize=(20, 10), title='LUBAX Beta')
```

```python
# Next we can calculate Sharpe ratios:
sharpe_ratios = ((analysis_df.mean() * 252) / (analysis_df.std() * np.sqrt(252)))
sharpe_ratios
```

```python
# Displaying the sharpe ratios as a bar chart:
sharpe_ratios.plot(kind='bar')
```

```python
# Lubax, due to its incerdibly low volatility, makes comparisons between them all moot. A more appropriate comparison would be between all of them except for LUBAX:
sharpe_ratios_better = ((no_lubax_df.mean() * 252) / (no_lubax_df.std() * np.sqrt(252)))
sharpe_ratios_better.plot(kind='bar')
```
